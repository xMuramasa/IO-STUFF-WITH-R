Funcion para separar datos en train y test con X del total en test

#### Training con 30% de los datos

```r
createData<-function(X, d, show){
    #Seteamos una semilla para replicar resultados
    set.seed(42)
    # En este caso utilizaremos un conjunto de entrenamiento con el X% de los datos
    train_size <- floor(X * nrow(d)) #El tamaño será el X% de la cantidad de filas
    train_mask <- sample(seq_len(nrow(d)),size= train_size) #Mascara aleatoria
    # Separamos los conjuntos
    train <- data[train_mask, ]
    test <- data[-train_mask, ]
    # se crea el arbol
    t = tree(Nota.Final ~ Sexo + Horas.Estudio.Semanal + VTR + Tiempo.Libre + Carrete + Salud + Inasistencias, data = train)
    if(show){
        # plot del arbol
        plot(t)
        text(t,pretty=1)
        # se muestra datos del arbol
        print(summary(t))
    }
    
    # variable retorno
    dat <- list("train" = train, "test" = test, "tree" = t)
    
    return(dat)
    }
```

#### Test con 30% de los datos

```r
newDat1 <- createData(X=0.3, d=data, show=T)
```

Matriz de Confusion

```r
# Predecimos cada tier para los datos de pruebas
test <- newDat1$test
pred <- predict(newDat1$tree,test,type='class')
# Creamos la matriz de confusión
conf_matrix1 <- with(test,table(pred,test$Nota.Final))
conf_matrix1
```

```r
# precision del modelo
acc <- sum(diag(conf_matrix1))/nrow(test)
miss_class_error1 <- 1 - acc
miss_class_error1
```

#### Test con 50% de los datos

```r
# training con 30% de los datos
newDat2 <- createData(0.5, data, show=T)
```

```r
# Predecimos cada tier para los datos de pruebas
test <- newDat2$test
pred <- predict(newDat2$tree,test,type='class')
# Creamos la matriz de confusión
conf_matrix2 <- with(test,table(pred,test$Nota.Final))
conf_matrix2
```
